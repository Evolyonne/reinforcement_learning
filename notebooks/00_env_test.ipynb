# Test de l’environnement BipedalWalker-v3

Objectif :
- vérifier le bon fonctionnement de l’environnement
- comprendre les espaces d’observation et d’action
- tester un agent aléatoire comme baseline naïve

```python
import numpy as np
import torch
import random

seed = 42
np.random.seed(seed)
torch.manual_seed(seed)
random.seed(seed)
```

```python
env = gym.make("BipedalWalker-v3", render_mode="human")
obs, info = env.reset(seed=seed)

print("Observation shape :", obs.shape)
print("Action space :", env.action_space)
print("Observation space :", env.observation_space)
```

```python
done = False
total_reward = 0.0
step = 0

while not done:
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)

    total_reward += reward
    done = terminated or truncated
    step += 1

print(f"Episode terminé en {step} steps")
print(f"Reward totale : {total_reward:.2f}")
env.close()
```

## Analyse

L’agent aléatoire échoue rapidement :
- chutes fréquentes
- récompense faible ou négative

Cela confirme :
- la complexité de l’environnement
- la nécessité d’algorithmes de RL continus (PPO, A2C, SAC)
